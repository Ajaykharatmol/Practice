```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\\New folder\\train_gbW7HTd.csv")
df
df["Workclass"].fillna("No Workclass", inplace = True) 
df["Occupation"].fillna("No Occupation", inplace = True) 
df["Native_Country"].fillna("No Native_Country", inplace = True)   
df.head()
```

```python
df1 = pd.read_csv(r"D:\\New folder\\test_2AFBew7.csv")
df1["Workclass"].fillna("No Workclass", inplace = True) 
df1["Occupation"].fillna("No Occupation", inplace = True) 
df1["Native_Country"].fillna("No Native_Country", inplace = True)   
#df1.head()
```

```python
inputs = df.drop('Income_Group',axis='columns')
target = df["Income_Group"]
```

```python
from sklearn.preprocessing import LabelEncoder
ID_A = LabelEncoder()
Age_A= LabelEncoder()
Workclass_A = LabelEncoder()
Education_A=LabelEncoder()
Marital_Status_A=LabelEncoder()
Occupation_A=LabelEncoder()
Relationship_A=LabelEncoder()
Race_A =LabelEncoder()
Sex_A=LabelEncoder()
Hours_Per_Week_A=LabelEncoder()
Native_Country_A=LabelEncoder()
```

```python
inputs['ID_n'] = ID_A.fit_transform(inputs['ID'])
inputs['Age_n'] = Age_A.fit_transform(inputs['Age'])
inputs['Workclass_n'] = Workclass_A.fit_transform(inputs['Workclass'])
inputs['Education_n'] = Education_A.fit_transform(inputs['Education'])
inputs['Marital_Status_n'] = Marital_Status_A.fit_transform(inputs['Marital_Status'])
inputs['Occupation_n'] = Occupation_A.fit_transform(inputs['Occupation'])
inputs['Relationship_n'] = Relationship_A.fit_transform(inputs['Relationship'])
inputs['Race_n'] = Race_A.fit_transform(inputs['Race'])
inputs['Sex_n'] = Sex_A.fit_transform(inputs['Sex'])
inputs['Hours_Per_Week_n'] = Hours_Per_Week_A.fit_transform(inputs['Hours_Per_Week'])
inputs['Native_Country_n'] = Native_Country_A.fit_transform(inputs['Native_Country'])
```

```python
inputs_n = inputs.drop(['ID','Age','Workclass','Education','Marital_Status','Occupation','Relationship','Race','Sex','Hours_Per_Week','Native_Country',],axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
ID_B = LabelEncoder()
Age_B= LabelEncoder()
Workclass_B = LabelEncoder()
Education_B=LabelEncoder()
Marital_Status_B=LabelEncoder()
Occupation_B=LabelEncoder()
Relationship_B=LabelEncoder()
Race_B =LabelEncoder()
Sex_B=LabelEncoder()
Hours_Per_Week_B=LabelEncoder()
Native_Country_B=LabelEncoder()
```

```python
df1['ID_n1'] = ID_B.fit_transform(df1['ID'])
df1['Age_n1'] = Age_B.fit_transform(df1['Age'])
df1['Workclass_n1'] = Workclass_B.fit_transform(df1['Workclass'])
df1['Education_n1'] = Education_B.fit_transform(df1['Education'])
df1['Marital_Status_n1'] = Marital_Status_B.fit_transform(df1['Marital_Status'])
df1['Occupation_n1'] = Occupation_B.fit_transform(df1['Occupation'])
df1['Relationship_n1'] = Relationship_B.fit_transform(df1['Relationship'])
df1['Race_n1'] = Race_B.fit_transform(df1['Race'])
df1['Sex_n1'] = Sex_B.fit_transform(df1['Sex'])
df1['Hours_Per_Week_n1'] = Hours_Per_Week_B.fit_transform(df1['Hours_Per_Week'])
df1['Native_Country_n1'] = Native_Country_B.fit_transform(df1['Native_Country'])
```

```python
inputs_n1 = df1.drop(['ID','Age','Workclass','Education','Marital_Status','Occupation','Relationship','Race','Sex','Hours_Per_Week','Native_Country',],axis='columns')
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n, target)
model.score(inputs_n,target)
model.predict([[0,22,7,9,4,0,1,4,1,39,39]])
```
